### 單根檢定 

```python

import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt

# 定義一個函數來進行ADF檢驗並顯示結果
def adf_test(series, stock_name):
    print(f'ADF Test for {stock_name}:')
    result = adfuller(series.dropna())  # 去掉缺失值
    labels = ['ADF Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used']
    for value, label in zip(result, labels):
        print(f'{label}: {value}')
    
    if result[1] <= 0.05:
        print(f" => {stock_name} 的數據沒有單根（也就是定態），拒絕單根假設（H0）\n")
    else:
        print(f" => {stock_name} 的數據具有單根（也就是非定態），無法拒絕單根假設（H0）\n")

# 對兩支股票進行ADF檢測
adf_test(tw_common['Adj Close'], '2465')
adf_test(us_common['Adj Close'], 'SMCI')

```

## Cointegration test 

### Johansen 驗證方法： 適用 較大樣本

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.vector_ar.vecm import coint_johansen

# 取出台股和美股的收盘价数据
tw_close = tw_common['Adj Close']
us_close = us_common['Adj Close']

# 合并两个时间序列为 DataFrame，Johansen 检验需要输入一个矩阵
data = pd.concat([tw_close, us_close], axis=1)
data.columns = ['TW_Close', 'US_Close']

# 确保没有 NaN 值，如果有缺失值可以进行填充或去除
data = data.dropna()

# Johansen 检验
# 参数解释：
# det_order = 0: 不包含截距项（你可以根据数据特性选择 0，1，2）
# k_ar_diff: 自回归滞后阶数，一般选择 1 到 2
johansen_test = coint_johansen(data, det_order=0, k_ar_diff=1)

# 检验结果输出
print("Eigenvalue statistic:", johansen_test.lr1)
print("Critical values (90%, 95%, 99%):", johansen_test.cvt)

# 查看各个统计量的解释
for i, eigen_stat in enumerate(johansen_test.lr1):
    print(f"Cointegration rank {i+1}:")
    print(f"Test Statistic: {eigen_stat}")
    print(f"Critical values at 90%, 95%, 99%: {johansen_test.cvt[i]}")

# 判断是否存在协整关系
# 如果 lr1 的统计量超过临界值，则表明存在协整关系
for i, eigen_stat in enumerate(johansen_test.lr1):
    if eigen_stat > johansen_test.cvt[i, 1]:  # 1 对应 95% 临界值
        print(f"At rank {i+1}, we reject the null hypothesis: There is cointegration.")
    else:
        print(f"At rank {i+1}, we fail to reject the null hypothesis: No cointegration.")


"""
exist conintegration in 1 rank 差分 : 表明它们之间有一个稳定的长期线性组合。

""" 
```

### Phillips–Ouliaris cointegration test 

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint

# 取出台股和美股的收盘价数据
tw_close = tw_common['Adj Close']
us_close = us_common['Adj Close']

# 确保数据没有缺失值
data = pd.concat([tw_close, us_close], axis=1)
data.columns = ['TW_Close', 'US_Close']
data = data.dropna()

# 使用 Phillips–Ouliaris 协整检验 (coint)
# coint() 返回 (t-statistic, p-value, critical_values)
po_test = coint(data['TW_Close'], data['US_Close'])

# 输出检验结果
print("Phillips-Ouliaris 检验统计量:", po_test[0])
print("p-value:", po_test[1])
print("临界值 (90%, 95%, 99%):", po_test[2])

# 基于 p-value 判断协整关系是否存在
if po_test[1] < 0.05:
    print("结果: 拒绝原假设，台股和美股之间存在协整关系")
else:
    print("结果: 无法拒绝原假设，台股和美股之间没有协整关系")


```

```python
import matplotlib.pyplot as plt

# 1. 計算收益率
tw_common['Return'] = tw_common['Adj Close'].pct_change().dropna()
us_common['Return'] = us_common['Adj Close'].pct_change().dropna()

# 2. 計算收益率差異（spread）
spread = tw_common['Return'] - us_common['Return']

# 3. 對 spread 進行 Z-score 標準化
spread_mean = spread.mean()
spread_std = spread.std()
spread_z = (spread - spread_mean) / spread_std



# 4. 繪製圖表
fig, ax1 = plt.subplots(figsize=(16, 8))

# 繪製收益率差異的 Z-score
ax1.plot(spread.index, spread_z, label='Spread (Return Difference Z-score)', color='red', linestyle='-', linewidth=2)
ax1.set_xlabel('Date')
ax1.set_ylabel('Standardized Return Spread (Z-score)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# 繪製成交量的標準化數據


# 設置標題
plt.title('Standardized Spread (Return Difference) Over Last Year')

# 顯示圖例
ax1.legend(loc='upper left')

plt.show()
```

### Long Only + Optuna threshold 

```python


# 1. Load Data (assumed tw_common['Adj Close'] and us_common['Adj Close'] are already available)
tw_close = tw_common['Adj Close']  # 台股收盘价
us_close = us_common['Adj Close']  # 美股收盘价

# 2. Define Objective Function for Optuna
def objective(trial):
    # Define the parameters to be optimized
    window = trial.suggest_int('window', 10, 100)  # Rolling window size between 10 and 100
    entry_threshold = trial.suggest_uniform('entry_threshold', 1, 3.0)  # Optimize entry threshold
    exit_threshold = trial.suggest_uniform('exit_threshold', 0, 1.0)  # Optimize exit threshold

    # 3. Calculate Spread and Z-Score Using a Rolling Window
    spread = tw_close - us_close
    rolling_mean = spread.rolling(window).mean()
    rolling_std = spread.rolling(window).std()

    # Calculate Z-score
    spread_z = (spread - rolling_mean) / rolling_std

    # 4. Generate Trade Signals Based on Z-Score Mean Reversion (Long-Only)
    long_entries = spread_z < -entry_threshold  # Z-score below -entry_threshold: long tw_close, short us_close
    long_exits = spread_z > -exit_threshold  # Z-score near 0: close the position

    # 5. Construct the Portfolio with Vectorbt (Long-Only Strategy)
    pf = vbt.Portfolio.from_signals(
        tw_close,  # Using tw_close as the base
        entries=long_entries,
        exits=long_exits,
        freq='d',  # Daily frequency
        fees=0.0014,  # Transaction fees
        slippage=0.001  # Slippage
    )
    
    # 6. Retrieve the Calmar ratio
    stats = pf.stats()
    calmar_ratio = stats['Calmar Ratio']  # Extract the Calmar ratio from the stats
    
    # Return the negative of the Calmar ratio (since Optuna minimizes the objective function)
    return -calmar_ratio

# 7. Run Optuna to optimize the rolling window and entry/exit thresholds
study = optuna.create_study(direction='minimize')  # Set direction to maximize Calmar ratio
study.optimize(objective, n_trials=400)  # Run 100 optimization trials

# 8. Get the best result
print(f"Best window: {study.best_params['window']}")
print(f"Best entry_threshold: {study.best_params['entry_threshold']}")
print(f"Best exit_threshold: {study.best_params['exit_threshold']}")
print(f"Best Calmar Ratio: {study.best_value}")

# 9. (Optional) Visualize the optimization history
optuna.visualization.plot_optimization_history(study).show()
```

```python
import vectorbt as vbt
import pandas as pd
import numpy as np
import optuna  # To perform optimization
from statsmodels.tsa.stattools import coint

# 1. Load Data (assumed tw_common['Adj Close'] and us_common['Adj Close'] are already available)
tw_close = tw_common['Adj Close']  # 台股收盘价
us_close = us_common['Adj Close']  # 美股收盘价

# 2. Define Objective Function for Optuna
def objective(trial):
    # Define the parameter to be optimized
    window = trial.suggest_int('window', 10, 100)  # Rolling window size between 10 and 100
    
    # 3. Calculate Spread and Z-Score Using a Rolling Window
    spread = tw_close - us_close
    rolling_mean = spread.rolling(window).mean()
    rolling_std = spread.rolling(window).std()

    # Calculate Z-score
    spread_z = (spread - rolling_mean) / rolling_std

    # 4. Define Entry/Exit Thresholds
    entry_threshold = 2  # Fixed entry threshold
    exit_threshold = 0.5  # Fixed exit threshold

    # 5. Generate Trade Signals Based on Z-Score Mean Reversion
    long_entries = spread_z < -entry_threshold  # Z-score below -entry_threshold: long tw_close, short us_close
    long_exits = spread_z > -exit_threshold  # Z-score near 0: close the position

    short_entries = spread_z > entry_threshold  # Z-score above entry_threshold: short tw_close, long us_close
    short_exits = spread_z < exit_threshold  # Z-score near 0: close the position

    # 6. Construct the Portfolio with Vectorbt
    pf = vbt.Portfolio.from_signals(
        tw_close,  # Using tw_close as the base
        entries=long_entries,
        exits=long_exits,
        short_entries=short_entries,
        short_exits=short_exits,
        freq='d',  # Daily frequency
        fees=0.0014,  # Transaction fees
        slippage=0.001  # Slippage
    )
    
    # 7. Retrieve the Calmar ratio
    stats = pf.stats()
    calmar_ratio = stats['Calmar Ratio']  # Extract the Calmar ratio from the stats
    
    # Return the negative of the Calmar ratio (since Optuna minimizes the objective function)
    return -calmar_ratio

# 8. Run Optuna to optimize the rolling window
study = optuna.create_study(direction='minimize')  # We minimize because we return -calmar_ratio
study.optimize(objective, n_trials=100)  # Run 50 optimization trials

# 9. Get the best result
print(f"Best window: {study.best_params['window']}")
print(f"Best Calmar Ratio: {-study.best_value}")

# 10. (Optional) Visualize the optimization history
optuna.visualization.plot_optimization_history(study).show()

```
