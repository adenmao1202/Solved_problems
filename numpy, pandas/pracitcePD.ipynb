### Chuksize : 分批load data 進來

- 你只需要分组统计的结果，并且希望逐步累积结果。
- 内存限制 使得你无法一次性加载整个数据集。

### Swifter 
- 處理大數據 --> 並行化處理 
- swifter 是一個用於加速 pandas 操作的 Python 庫。它自動化地決定應該使用普通的 pandas.apply()，還是並行處理（並行化）來加速 apply() 操作。這對於在大型數據集上進行操作時特別有用。

- 加速 apply function 

```python
import pandas as pd
import swifter

# 假設我們有一個大的 DataFrame
df = pd.DataFrame({
    'a': range(1000000),
    'b': range(1000000, 2000000)
})

# 使用 swifter 加速 apply 操作
df['c'] = df.swifter.apply(lambda x: x['a'] +  x['b'], axis=1)
df
```

```python

# 使用 swifter 加速 applymap 操作
df = df.swifter.applymap(lambda x: x * 2)
df
```

```python
df = df.applymap(lambda x: x * 2)
df
```

### concat vs. merge

- merge: on = 靠什麼合併，  how  =  如何合併？

```python

# Create two DataFrames
df1 = pd.DataFrame({'key': ['A', 'B', 'C'], 'value1': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['B', 'C', 'D'], 'value2': [4, 5, 6]})

# Merge on 'key' column
inner_df = pd.merge(df1, df2, on='key', how='inner')
inner_df
```

```python
outer_df = pd.merge(df1, df2, on='key', how='outer')
outer_df
```

- concat: 單純的把資料串起來 

```python

# Create two DataFrames
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [7, 8, 9], 'B': [10, 11, 12]})

# Concatenate DataFrames vertically (row-wise)
concat_df = pd.concat([df1, df2], axis = 0)    # 以row 的方式串接
concat_df
```

```python
concat = pd.concat([df1, df2], axis = 1)    # 以 col 的方式 串接
concat
```
